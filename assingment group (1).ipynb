```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

base_url = "https://www.scholars4dev.com/tag/scholarships-for-rwandans/page/{}/"
all_scholarships = []

def fetch_page(url, retries=3, timeout=10):
    for i in range(retries):
        try:
            response = requests.get(url, timeout=timeout)
            response.raise_for_status()
            return response
        except requests.exceptions.RequestException as e:
            print(f"Attempt {i+1} failed for {url}: {e}")
            time.sleep(5)
    return None

for page in range(1, 11):  
    url = base_url.format(page)
    print(f"Scraping page {page}...")

    response = fetch_page(url)
    if not response:
        print(f"Failed to retrieve page {page}. Skipping...")
        continue

    soup = BeautifulSoup(response.content, "html.parser")
    posts = soup.find_all('div', class_='post clearfix')
    
    if not posts:
        print(f"No posts found on page {page}. Stopping scraper.")
        break
    
    for post in posts:
        try:
            h2_tag = post.find('h2')
            a_tag = h2_tag.find('a') if h2_tag else None
            title = a_tag.text.strip() if a_tag else None
            link = a_tag['href'] if a_tag else None
            
            org_div = post.find_all('div', class_='post_column_1')
            sponsor = None
            if len(org_div) > 0:
                em_tag = org_div[0].find('em')
                sponsor = em_tag.text.strip() if em_tag else org_div[0].get_text(strip=True)
            
            deadline = None
            if len(org_div) > 1:
                text = org_div[1].get_text(separator=" ", strip=True)
                if "Deadline:" in text:
                    deadline = text.split("Deadline:")[1].split("Study in:")[0].strip()
            
            if title and link:
                all_scholarships.append({
                    "Title": title,
                    "Link": link,
                    "Sponsor": sponsor,
                    "Deadline": deadline
                })
            else:
                print("Skipped one post - missing title or link.")
        except Exception as e:
            print(f"Error parsing post: {e}")
    
    time.sleep(3)  # polite delay

df = pd.DataFrame(all_scholarships)
print(f"Scraped {len(df)} scholarships total.")
df.to_csv("scholarships_rwandans.csv", index=False)
print("Saved data to scholarships_rwandans.csv")
```
